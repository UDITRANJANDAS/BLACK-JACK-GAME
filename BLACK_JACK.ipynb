# BLACK JACK GAME

```python

import random

# creating the deck of cards
card_categories = ['Hearts', 'Diamonds', 'Clubs', 'Spades']
cards_lists = ['Ace', '2', '3', '4', '5', '6', '7', '8', '9', '10', 'Jack', 'Queen', 'King']
deck = [(card, category) for category in card_categories for card in cards_lists]

# FUNCTION TO ASSIGN POINTS
def card_value(card):
    if card[0] in ['Jack', 'Queen', 'King']:
        return 10
    elif card[0] == 'Ace':
        return 11
    else:
        return int(card[0])

# shuffle the deck
random.shuffle(deck)

# initial cards
player_card = [deck.pop(), deck.pop()]
dealer_card = [deck.pop(), deck.pop()]

# game loop
while True:
    player_score = sum(card_value(card) for card in player_card)
    dealer_score = sum(card_value(card) for card in dealer_card)

    print("\nCards Player has:", player_card)
    print("Score of the Player:", player_score)

    # check if player busted
    if player_score > 21:
        print("\nPlayer exceeded 21! Dealer wins.")
        break

    choice = input('What do you want? ["play" to draw, "stop" to hold]: ').lower()

    if choice == "play":
        new_card = deck.pop()
        player_card.append(new_card)

    elif choice == "stop":
        break

    else:
        print("Invalid choice. Please try again.")

# dealer logic (simple)
while dealer_score < 17:
    dealer_card.append(deck.pop())
    dealer_score = sum(card_value(card) for card in dealer_card)

# final results
print("\nFinal Results:")
print("Player cards:", player_card, "| Score:", player_score)
print("Dealer cards:", dealer_card, "| Score:", dealer_score)

if dealer_score > 21:
    print("Dealer exceeded 21! Player wins.")
elif player_score > dealer_score:
    print("Player wins!")
elif player_score < dealer_score:
    print("Dealer wins!")
else:
    print("It's a tie!")
```
